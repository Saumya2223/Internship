# Q1

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

table = soup.find('table', {'class': 'wikitable sortable'})
rows = table.find_all('tr')[1:]

rank_list = []
name_list = []
artist_list = []
upload_date_list = []
views_list = []

for row in rows:
    cols = row.find_all('td')
    if len(cols) < 5:
        continue
        
    rank = cols[0].text.strip()
    rank_list.append(rank)
    
    name = cols[1].text.strip()
    name_list.append(name)
    
    artist = cols[2].text.strip()
    artist_list.append(artist)
    
    upload_date = cols[4].text.strip()
    upload_date_list.append(upload_date)
    
    views = cols[3].text.strip()
    views_list.append(views)

name_list1=[]
for i in name_list:
    name_list1.append(i.split('[')[0])
    
df = pd.DataFrame({'Rank': rank_list, 'Name': name_list1, 'Artist': artist_list, 'Upload Date': upload_date_list, 'Views': views_list})
print(df.to_string(index=False))
```
