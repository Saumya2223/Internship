```python
# Initialize the WebDriver for the Microsoft Edge browser
drv=webdriver.Edge(r"C:\Users\Rahul\Desktop\Internship\edgedriver_win64\msedgedriver")

#Opening Flipkart.com for Scraping
drv.get('https://www.flipkart.com/')

#Closing the login screen
clog=drv.find_element(By.XPATH,'/html/body/div[2]/div/div/button')
clog.click()

#Inputing Product as Sunglasses in search box
searchinput=drv.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')
searchinput.send_keys('Sunglasses')

# Find the search button using absolute XPATH and click it
searchclick=drv.find_element(By.CLASS_NAME,'_34RNph')
searchclick.click()

#Creating lists to store the details
Brand=[]
prd=[]
price=[]

#No. of search results is 40 results per page, requirement is 100 results. Therefore we need to scrape multiple pages.
#Using for loop for navigating and scraping data from multiple pages.

for page in range(4):
    brand1=drv.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    for i in brand1:
        Brand.append(i.text)
    pr1=drv.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
    for i in pr1:
        prd.append(i.text)
    pric1=drv.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    for i in pric1:
        price.append(i.text)
    time.sleep(2)
    #As class of NEXT buttong and PREVIOUS button is same, targeting the "Next" button using its child text nodes
    next_button=drv.find_element(By.XPATH, '//a[contains(., "Next")]')
    next_button.click()
    time.sleep(1)

#Making Lists of first 100 elements
Brand100=Brand[0:100]
Prod100=prd[0:100]
Price100=price[0:100]

#Creating a Pandas DataFrame
dfr=pd.DataFrame({'Brand':Brand100,'Product Description':Prod100,'Price':Price100})
dfr
```
