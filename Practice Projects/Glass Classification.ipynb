Standardization of data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
x
```

Replacing values in Type of glass as a binary for classification.

*If type of class = 0 then window glass, if type of glass = 1 then not window glass.

```python
def classify_glass(val):
    if val in [1,2,3,4]:
        return '0'
    else:
        return '1'
y=y.apply(lambda x: classify_glass(x))
```

```python
y = y.astype('int')
```

```python
y
```

# Modeling

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
maxAccu=0
maxRS=0

for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=i)
    rfr=RandomForestClassifier()
    rfr.fit(x_train, y_train)
    pred=rfr.predict(x_test)
    acc=accuracy_score(y_test,pred)
    if acc>maxAccu:
        maxAccu = acc
        maxRs = i
print('Best acc is ', maxAccu, 'at random state', maxRs)
```

Best accuracy of the model is 100% at random state = 5.

```python
#importing more models
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier,AdaBoostClassifier, BaggingClassifier
from sklearn.metrics import classification_report,confusion_matrix,roc_curve,accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
```

Random Forest Classifier model

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20,random_state=maxRs)
rfc=RandomForestClassifier()
rfc.fit(x_train, y_train)
prfc=rfc.predict(x_test)
print(accuracy_score(y_test,prfc))
print(confusion_matrix(y_test,prfc))
print(classification_report(y_test,prfc))
```

The model Accuracy is 97%.

Logistic regression model

```python
lr=LogisticRegression()
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20,random_state=maxRs)
lr.fit(x_train, y_train, sample_weight=None)
plr=lr.predict(x_test)
print(accuracy_score(y_test,plr))
print(confusion_matrix(y_test,plr))
print(classification_report(y_test,plr))
```

The model Accuracy is 95%.

```python
svc=SVC()
svc.fit(x_train,y_train)
psvc=svc.predict(x_test)
print(accuracy_score(y_test,psvc))
print(confusion_matrix(y_test,psvc))
print(classification_report(y_test,psvc))
```

The model Accuracy is 97%.

The model Accuracy is 97%.

```python
#importing RandomizedseachCV for hyperparameter tuning
from sklearn.model_selection import RandomizedSearchCV
```

```python
#parameters for Bagging classifier model:
parameters = {'n_estimators': [50, 100, 150],
              'loss':['log_loss','deviance','exponential'],
              'criterion':['friedman_mse', 'squared_error'],
              'learning_rate': [0.1, 0.05, 0.01],
              'max_depth': [2, 3, 4],
              'min_samples_split': [2, 5, 10],
              'min_samples_leaf': [1, 2, 4],
              'max_features': ['sqrt', 'log2'],
              'subsample': [0.6, 0.8, 1.0],
              'random_state': [10, 20, 30],
              'verbose':[0,1],
              'warm_start':[True,False],
              'max_leaf_nodes':[2,3,5]
             }
```

```python
rscv = RandomizedSearchCV(GradientBoostingClassifier(),parameters,cv=5)
rscv.fit(x_train,y_train)
```

```python
#Finding the best parameters
rscv.best_params_
```

```python
#Creating the final model
Final_model=GradientBoostingClassifier(warm_start=False,
 verbose= 1,
 subsample=0.6,
 random_state=20,
 n_estimators=100,
 min_samples_split=2,
 min_samples_leaf=4,
 max_leaf_nodes=5,
 max_features='log2',
 max_depth=2,
 loss='log_loss',
 learning_rate=0.1,
 criterion='friedman_mse'
)
```

```python
#Training the final model
Final_model.fit(x_train,y_train)
```

```python
#Final prediction
Fp=Final_model.predict(x_test)
acc=accuracy_score(y_test,Fp)
print(acc*100)
```

Final model accuracy is 97%.

```python
y_pred_prob = Final_model.predict_proba(x_test)[:,1]
fpr,tpr,thresholds=roc_curve(y_test,y_pred_prob)
roc_auc= auc(fpr,tpr)

plt.plot([0,1],[0,1], 'k--')
plt.plot(fpr,tpr,label='GradientBoostingClassifier(area=%0.2f)'%roc_auc)
plt.show()
auc_score=roc_auc_score(y_test,Final_model.predict(x_test))
print('Auc Score: ','\n',auc_score)
```
