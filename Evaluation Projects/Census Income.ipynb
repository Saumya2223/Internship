# Modeling

```python
#importing models and libraries
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier,AdaBoostClassifier, BaggingClassifier
from sklearn.metrics import classification_report,confusion_matrix,roc_curve,accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
maxAccu=0
maxRS=0

for i in range(1,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=i)
    lr=LogisticRegression()
    lr.fit(x_train, y_train)
    pred=lr.predict(x_test)
    acc=accuracy_score(y_test,pred)
    if acc>maxAccu:
        maxAccu = acc
        maxRs = i
print('Best acc is ', maxAccu, 'at random state', maxRs)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=i)
def model(m):
    m.fit(x_train,y_train)
    predt=m.predict(x_train)
    pred=m.predict(x_test)
    print('Training Accuracy Score: ',accuracy_score(y_train,predt)*100)
    print('Accuracy Score: ',accuracy_score(y_test,pred)*100)
    print(confusion_matrix(y_test,pred))
    print(classification_report(y_test,pred))
```

# Gradient Boosting Classifier (GBC) is the best model!

Hyperparameter tuning

```python
#importing RandomizedseachCV for hyperparameter tuning
from sklearn.model_selection import RandomizedSearchCV
```

```python
parameters = {
    'loss':['log_loss','deviance','exponential'],
    'learning_rate':[0.0,0.1,0.5],
    'n_estimators': [50, 100],
    'criterion':['friedman_mse','squared_error'],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'max_depth': [2, 3, None],
    'max_features': ['sqrt', 'log2','auto'],
    'random_state': [10, None,22],
    'warm_start':[True,False]
}
```

```python
rsv = RandomizedSearchCV(GradientBoostingClassifier(), parameters)
rsv.fit(x_train, y_train)
print(rsv.best_params_)
```

```python
Hyper_gbc=GradientBoostingClassifier(loss='exponential',learning_rate=0.1,n_estimators=100,criterion='squared_error',min_samples_split=2,min_samples_leaf=2,max_depth=None,max_features='log2',warm_start=True,random_state=22)
```

```python
model(Hyper_gbc)
```

# Accuracy after hyperparameter tuning of Gradient Boosting Classifier is 88.76%

```python
from sklearn.metrics import roc_curve,roc_auc_score
y_pred_prob=Hyper_gbc.predict_proba(x_test)[:,1]
fpr,tpr,thersholds=roc_curve(y_test,y_pred_prob)
print('ROC Score',roc_auc_score(y_test,Hyper_gbc.predict(x_test)))
```
